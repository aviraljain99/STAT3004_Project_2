```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import exponential as exp
%matplotlib notebook
```

```python
I_0 = 5
S_0 = 5
N = 10
Ris = 2
```

```python
def getQMatrix(Rsi):
    # Create Q matrix
    Q = np.zeros((N + 1, N + 1))
    for i in range(N):
        Q[i, i + 1] = Ris * (N - i) # infective becomes a susceptible again
        if i != 0:
            Q[i, i - 1] = Rsi * i * (N - i) # susceptible becomes an infective
        Q[i, i] = - np.sum(Q[i, :])
    return Q

print(getQMatrix(2))
```

```python
Rs = np.linspace(0, 1, num=40)
probs = []
trials = int(500)

for Rsi in Rs:
    print(Rsi)
    # Get Q matrix
    Q = getQMatrix(Rsi)
    diagonal = (- np.diag(Q))
    # Get probability matrix
    Pjump = Q / diagonal[:, np.newaxis]
    np.fill_diagonal(Pjump, 0)
    np.nan_to_num(Pjump, copy=False)
    
#     with np.printoptions(precision=2):
#         print(Pjump)
    event = 0
    for trial in range(trials):
        # print(trial)
        time = 0
        state = S_0
        # print(state)
        while state != N:
            time += exp(scale=1/diagonal[state])
            state = np.random.choice(range(0, N + 1), p=Pjump[state, :])
            # print(state)
        if time > 15:
            event += 1
        # print("")
    print(event)
    probs.append(event/trials)

print(event)
```

```python
plt.figure()
plt.scatter(Rs, probs)
plt.plot(Rs, np.ones(len(probs)) * 0.9)
plt.xlabel("Rsi value")
plt.ylabel("Probability")
plt.savefig("q4_part_a")
```
